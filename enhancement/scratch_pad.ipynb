```python

def amp_module(img, m):
  """
  img.shape - (3, H, W)
  m - central value (typically 0.5) 
  """
  H = img.shape[0]
  W = img.shape[1]
  sumOfPixels = torch.sum(img)
  amp = m*(3*H*W)/(sumOfPixels)
  return amp
```
